#### Converting runnabales to standard component and chain together 

```python
from abc import ABC, abstractmethod
import random
```

```python
class Runnable(ABC):

  @abstractmethod
  def invoke(input_data):
    pass
```

```python
class NakliLLM(Runnable):

  def __init__(self):
    print('LLM created')

  def invoke(self , prompt):
      response_list = [
        'delhi is the capital of india',
        'ipl is the indian league',
        'ai stands for the artificial intelligence'
      ]

      return { 'response': random.choice(response_list)}

## remain this class for old method
  def predict(self , prompt):

    response_list = [
      'delhi is the capital of india',
      'ipl is the indian league',
      'ai stands for the artificial intelligence'
    ]

    return { 'response': random.choice(response_list)}
```

```python
class NakliPromptTemplate(Runnable):

  def __init__(self , template , input_variables) :
    self.template = template
    self.input_variables = input_variables


  def invoke(self , input_dict):  ## fill the template blanck
    return self.template.format(**input_dict)   
  
  
  def format(self , input_dict):
    return self.template.format(**input_dict)
```

```python
class NakliStrOutputparser(Runnable):
  def __init__(self):
    pass

  def invoke(self, input_data):  ## give only the ans resonse value nikal k
    return input_data['response']
```

```python
class RunnableConnector(Runnable):
  def __init__(self, runnable_list):
    self.runnable_list = runnable_list
  
  def invoke(self, input_data):

    for runnable in self.runnable_list:  # Har step ka output next step ka input ban raha hai
      input_data = runnable.invoke(input_data)

    return input_data
```

```python
template = NakliPromptTemplate(
  template = 'write a {length} poem about {topic}',
  input_variables= ['lenght', 'topic']
)
```

```python
llm = NakliLLM()
```

```python
parser = NakliStrOutputparser()
```

```python
chain = RunnableConnector([template , llm , parser])
```

```python
chain.invoke({'length':'long', 'topic':'india'})
```

```python
template1 = NakliPromptTemplate(
  template = 'Write a joke about {topic}',
  input_variables= ['topic']
)
```

```python
template2 = NakliPromptTemplate(
  template = 'Explaint the following joke {response}',
  input_variables= ['response']
)
```

```python
llm = NakliLLM()
```

```python
parser = NakliStrOutputparser()
```

```python
chain1 = RunnableConnector([template1 , llm])
```

```python
chain2 = RunnableConnector([template2,llm,parser])
```

```python
final_Chain = RunnableConnector([chain1 , chain2])
```

```python
final_Chain.invoke({'topic':'cricket'})
```
